```python
import matplotlib.pyplot as plt

#Datos
x = [1, 2, 3, 4, 5] #Colocamos los datos del eje x
y = [2, 3, 5, 7, 11] #Definimos los números primos en el eje y

#Creamos un plot
plt.plot(x,y, label = 'Prime Numbers', color = 'green', marker = '+') #Al crear el plot definimos la leyenda, el color verde en este caso y el tipo de marcador de cruz

#Añadimos título y leyendas
plt.title('Linea simple graficada') #Definimos título con .title
plt.xlabel('X Axis') #Definimos el eje X
plt.ylabel('Y Axis') #Definimos el eje y

#Mostrar leyenda
plt.legend()

#Imprimir la gráfica
plt.show()
```

```python
import matplotlib.pyplot as plt

#Datos
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

#Crear una figura con cuadrícula de 2x2 subgráficos
fig, axs = plt.subplots(2, 2)

#Gráfico de línea
axs[0, 0].plot(x,y,color='blue',marker='+')
axs[0, 0].set_title("Gráfico de línea")

#Diagrama de barras
axs[0, 1].bar(x,y,color='yellow')
axs[0, 1].set_title("Gráfico de barras")

#Diagrama de dispersión
axs[1, 0].scatter(x,y,color='green',marker='+')
axs[1, 0].set_title("Gráfico de dispersión")

#Histograma
axs[1, 1].hist(y,bins=10,color='red')
axs[1, 1].set_title("Histograma")

#Ajustamos el espacio entre subgráficos
plt.tight_layout()

#Imprimimos los subgráficos
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#Definir la figura y el eje en 3D
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')

#Generar datos y el gráfico de superficie
#Crear una cuadrícula

x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
X,Y=np.meshgrid(x,y)
#Definir una función z=f(x,y)
Z=np.sin(np.sqrt(X**2+Y**2))
#Crear el gráfico de superficie
ax.plot_surface(X,Y,Z,cmap='viridis')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
plt.show()
```
